```python
total_cols = 4
total_rows = len_num_var//total_cols # len_num_var//total_cols+1 may needed sometime
fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                        figsize=(7*total_cols, 7*total_rows), constrained_layout=True)
for i, var in enumerate(num_var):
    row = i//total_cols
    pos = i % total_cols
    
    sns.histplot(df[var], bins=20, kde = True, color = 'tab:olive', ax=axs[row][pos])
    sns.histplot(df3_drop_rows[var], bins=20, kde = True, ax=axs[row][pos])
```
