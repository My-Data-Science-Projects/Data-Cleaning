## Missing Value Imputation - Categorical Variable

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set_style("darkgrid")
warnings.filterwarnings('ignore')
```

```python
# load dataset
df = pd.read_csv(r"C:\Users\akash\jupyter notebook\ML_Indian_AI_Course\Data Cleaning\datasets\house-prices-advanced-regression-techniques\train.csv")
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
cat_vars = df.select_dtypes(include='object')
```

```python
cat_vars.head()
```

```python
miss_val_per = cat_vars.isnull().mean() * 100
miss_val_per
```

```python
drop_vars = ['Alley','FireplaceQu','PoolQC','Fence','MiscFeature']
cat_vars.drop(columns = drop_vars, axis = 1, inplace = True)
cat_vars.shape
```

```python
isnull_per = cat_vars.isnull().mean() * 100
miss_vars = isnull_per[ isnull_per > 0 ].keys()
miss_vars
```

```python
cat_vars['MasVnrType'].mode()

# cat_vars['MasVnrType'].fillna('---put any desire value---')
```

```python
cat_vars['MasVnrType'].value_counts()
```

```python
cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].mode()[0]) 
# here [0] to access 0th row from series, cause, we got None in 0th row after mode()
```

```python
for var in miss_vars:
    cat_vars[var].fillna(cat_vars[var].mode()[0], inplace = True) 
    print(var,"=",cat_vars[var].mode()[0])
```

```python
cat_vars.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(miss_vars):
    plt.subplot(4,3,i+1)
    plt.hist(cat_vars[var], label="Impute")
    plt.hist(df[var].dropna(), label="Original")
    plt.legend()
```

```python
df.update(cat_vars)
df.drop(columns=drop_vars, inplace=True)
```
