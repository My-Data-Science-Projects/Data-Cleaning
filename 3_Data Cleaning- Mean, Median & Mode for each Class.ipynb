```python
df['LotConfig'].unique()
```

```python
# Here take 'Inside' category and show only 'LotFrontage' value according to 'Inside' category of 'LotConfig' data
df[df.loc[:, 'LotConfig'] == 'Inside']['LotFrontage']
```

```python
# Here replace above null values

df[df.loc[:, 'LotConfig'] == 'Inside']['LotFrontage'].replace(np.nan, df[df.loc[:, 'LotConfig'] == 'Inside']['LotFrontage'].mean())
```

```python
# Fill null/NaN value with Mean by Class
df_copy = df.copy()

for var_class in df['LotConfig'].unique():
    df_copy.update(df[df.loc[:, 'LotConfig'] == var_class]['LotFrontage'].replace(np.nan, df[df.loc[:, 'LotConfig'] == var_class]['LotFrontage'].mean()))
```

```python
df_copy.isnull().sum()
```

```python
# Fill null/NaN value with Mean by Class
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig', 'MasVnrType', 'GarageType']

for cat_var, num_var_miss in zip(cat_vars, num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:, cat_var] == var_class][num_var_miss].replace(np.nan, df[df.loc[:, cat_var] == var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

#### In above, after run our for loop, there still nan exist in MasVnrType and GarageBlt because of nan values was present in their categorical column. When we want to fill some numerical column's null value by it's categorical variable, then categorical variables must be not null.

```python
df_copy[df_copy[['MasVnrType']].isnull().any(axis=1)]
```

```python
df_copy[df_copy[['GarageType']].isnull().any(axis=1)]
```

```python
# Fill null/NaN value with Mean by Class
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig', 'Exterior2nd', 'KitchenQual']

for cat_var, num_var_miss in zip(cat_vars, num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:, cat_var] == var_class][num_var_miss].replace(np.nan, df[df.loc[:, cat_var] == var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

## In above, we changed our categorical variables to deal with missing values in corresponding numerical variables. Here we didn't use domain knowledge.

# Data Distribution

```python
plt.figure(figsize=(20,20))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[var], bins=20, kde = True, color = 'tab:red', label = 'Original')
    sns.histplot(df_copy[var], bins=20, kde = True, color = 'tab:green', label = 'Mean')
    plt.legend()
```

# Median Method

```python
# Fill null/NaN value with Mean by Class
df_copy_median = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig', 'Exterior2nd', 'KitchenQual']

for cat_var, num_var_miss in zip(cat_vars, num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy_median.update(df[df.loc[:, cat_var] == var_class][num_var_miss].replace(np.nan, df[df.loc[:, cat_var] == var_class][num_var_miss].median()))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```

```python
# ORIGINAL - MEAN - MEDIAN

plt.figure(figsize=(20,20))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[var], bins=20, kde = True, color = 'tab:red', label = 'Original')
    sns.histplot(df_copy[var], bins=20, kde = True, color = 'tab:green', label = 'Mean')
    sns.histplot(df_copy_median[var], bins=20, kde = True, color = 'tab:blue', label = 'Median')
    plt.legend()
```
